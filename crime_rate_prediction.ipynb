### Model selection and Model building

```python
#Disturbing data into dependent and independent varaibales
x = data.drop(["crime_rate"], axis = 1) #independent data
y=data["crime_rate"] #dependant data
```

```python
x.head()
```

```python
y.head()
```

#### Train Test Splitting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15)
```

```python
x_train.head().sum()
```

```python
y_train.head().sum()
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)  #fit allows model to learn
```

```python
y_pred_lr = lr.predict(x_test)#predict() will help mlm prediction
```

```python
score = r2_score(y_test, y_pred_lr)
print(score)
```
